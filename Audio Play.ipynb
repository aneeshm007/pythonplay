```python
phone_char_tone_dict = {"1": (697, 1209), 
                        "2": (697, 1336), 
                        "3": (697, 1477),
                        "4": (770,1209),
                        "5": (770,1336),
                        "6": (770,1477),
                        "7": (852,1209),
                        "8": (852,1336),
                        "9": (852,1477),
                        "0": (941, 1336),
                        "*": (941,1209), 
                        "#": (941,1477),
                        "-": (0,0)}
```

```python
np.zeros(5)
```

```python
def audio_from_string(some_string, 
                      num_seconds=0.2, 
                      sample_rate=5000, 
                      pause_seconds = 0):
    audio_array = np.empty(len(some_string)*(int(np.ceil(num_seconds*sample_rate))+int(np.ceil(pause_seconds*sample_rate))))
    for phone_char_index in range(len(some_string)):
        if some_string[phone_char_index] not in phone_char_tone_dict.keys():
            raise KeyError("No DTMF tone exists for one of your chars")
        t = np.linspace(0, num_seconds, int(np.ceil(num_seconds*sample_rate)))
        the_tones = phone_char_tone_dict[some_string[phone_char_index]]
        x = np.cos(2*np.pi*the_tones[0]*t) + np.cos(2*np.pi*the_tones[1]*t)
        zeros_array = np.zeros((int(np.ceil(num_seconds*sample_rate))+int(np.ceil(pause_seconds*sample_rate))))
        zeros_array[0:int(np.ceil(num_seconds*sample_rate))] = x
        audio_array[(int(np.ceil(num_seconds*sample_rate))+int(np.ceil(pause_seconds*sample_rate)))*phone_char_index:(int(np.ceil(num_seconds*sample_rate))+int(np.ceil(pause_seconds*sample_rate)))*(phone_char_index+1)] = zeros_array 
    write("some_audio.wav", sample_rate, audio_array) 
```

```python
audio_from_string("1 5 3 3 3 - 3 5 3 1 1 - 1 5 3 6 # - # 9 6 3 ".replace(" ",""),num_seconds = 0.17, pause_seconds=0.1)
```
